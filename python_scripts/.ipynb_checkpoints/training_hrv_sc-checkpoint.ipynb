# Pre-processing Data and Feature Extraction
## Separating into sections

```python
labels = []
df["Section_labels"] = None

for i, g in df.groupby([(df.Stressed != df.Stressed.shift()).cumsum()]):
    label = g.Stressed.tolist()[0]
    labels.append(label)    
    df.at[g.index[0]: g.index[-1], "Section_labels"] = str(i) + '_' + str(label)

unique_sections = df["Section_labels"].unique()
unique_sections_num = df["Section_labels"].nunique()
print(unique_sections, unique_sections_num)
print(labels)
```

```python
label_count = {}
for label in labels:
    if label not in label_count:
        label_count[label] = 1
    else:
        label_count[label] += 1
        
print(label_count)
```

## Min-Max Normalisation

```python
def min_max_normalisation(df):
    min_val_gsr = min(df['Conductance (uS)'])
    max_val_gsr = max(df['Conductance (uS)'])
    scaling_gsr = max_val_gsr-min_val_gsr
    df.loc[:, 'Normalised_GSR'] = (df.loc[:, 'Conductance (uS)']- min_val_gsr)/scaling_gsr 
    
    min_val_hr = min(df["Heart_Rate"])
    max_val_hr = max(df["Heart_Rate"])
    scaling_hr = max_val_hr-min_val_hr
    df.loc[:, "Normalised_HR"] = (df.loc[:, "Heart_Rate"]- min_val_hr)/scaling_hr
    df = df.dropna()
    return df
```

```python
df = df.groupby(df["Section_labels"]).apply(min_max_normalisation)
```

```python
df
```

## Extracting data

```python
def find_rmssd(data):
    rr_diff = np.diff(data)
    rr_diff[-1] = 0
    rr_sqdiff = np.power(rr_diff, 2)
    rmssd = np.sqrt(np.mean(rr_sqdiff))
    return rmssd
```

```python
# labels already defined as "labels"
HR_means = []        # HR average value - Normalised data
HR_stds = []         # HR standard devation - Normalised data
HR_diffs = []        # Difference between HR Max and HR Min - Raw data

HR_HRV_means = []    # HR HRV average value (HRV) - Raw data
HR_HRV_stds = []     # HR HRV standard deviation (Variation of HRV) - Raw data
HR_HRV_RMSSDs = []   # HR HRV RMSSD  (Changes in HRV) - Raw data

GSR_means = []       # GSR average value (Skin Conductance Level) - Normalised data
GSR_stds = []        # GSR standard deviation (Skin Conductance Level variability) - Normalised data

GSR_num_peaks = []   # GSR peaks (Skin Conductance Responses) - Normalised data
GSR_rate_peaks = []  # GSR peaks over time (Skin Conductance Responses per second) - Normalised data
GSR_mean_peaks = []  # GSR peaks mean (Skin Conductance Response average) - Normalised data
GSR_std_peaks = []   # GSR peak std (Skin Conductance Response variability) - Normalised data
```

```python
for i in range(unique_sections_num):
    grouped = df.groupby(df["Section_labels"])
    section = grouped.get_group(unique_sections[i])
    time = section['Time (s)'].to_numpy()
    
    HR_data = section["Heart_Rate"].to_numpy()
    GSR_data = section["Conductance (uS)"].to_numpy()
    HR_norm_data = section["Normalised_HR"].to_numpy()
    GSR_norm_data = section["Normalised_GSR"].to_numpy()
    HRV_data = section["HRV"].to_numpy()
    
    HR_means.append(np.mean(HR_norm_data))
    HR_stds.append(np.std(HR_norm_data))
    HR_diffs.append(max(HR_data) - min(HR_data))
    
    HR_HRV_means.append(np.mean(HRV_data))
    HR_HRV_stds.append(np.std(HRV_data))
    HR_HRV_RMSSDs.append(find_rmssd(HRV_data))
    
    GSR_means.append(np.mean(GSR_norm_data))
    GSR_stds.append(np.std(GSR_norm_data))
    
    curr_peaks_ind, _ = find_peaks(GSR_norm_data)
    curr_peaks = GSR_norm_data[curr_peaks_ind]
    GSR_num_peaks.append(len(curr_peaks))
    GSR_rate_peaks.append(GSR_num_peaks[-1]/len(GSR_norm_data))
    GSR_mean_peaks.append(np.mean(curr_peaks))
    GSR_std_peaks.append(np.std(curr_peaks))
```

```python
stress_df = pd.DataFrame()
stress_df['HR_means'] = pd.Series(HR_means)
stress_df['HR_stds'] = pd.Series(HR_stds)
stress_df['HR_diffs'] = pd.Series(HR_diffs)
stress_df['HR_HRV_means'] = pd.Series(HR_HRV_means)
stress_df['HR_HRV_stds'] = pd.Series(HR_HRV_stds)
stress_df['HR_HRV_RMSSDs'] = pd.Series(HR_HRV_RMSSDs)
stress_df['GSR_means'] = pd.Series(GSR_means)
stress_df['GSR_stds'] = pd.Series(GSR_stds)
#stress_df['GSR_num_peaks'] = pd.Series(GSR_num_peaks)
stress_df['GSR_rate_peaks'] = pd.Series(GSR_rate_peaks)
stress_df['GSR_mean_peaks'] = pd.Series(GSR_mean_peaks)
stress_df['GSR_std_peaks'] = pd.Series(GSR_std_peaks)
stress_df['Stress'] = pd.Series(labels)

stress_df=stress_df.dropna()
stress_df
```

```python
def plot_data(x_name, y_name, data, labels):
    x_data = data.loc[:, [x_name]]
    y_data = data.loc[:, [y_name]]
    
    figure, ax1 = plt.subplots()
    scatter = ax1.scatter(x_data, y_data, c=labels, cmap="viridis")
    plt.title(y_name + " vs " + x_name)
    legend1 = ax1.legend(*scatter.legend_elements(), loc="upper right", title="Label")
    ax1.add_artist(legend1)
    plt.show()
```

```python
plot_data('HR_means', 'HR_stds', stress_df, labels)
plot_data('HR_means', 'HR_diffs', stress_df, labels)
plot_data('HR_HRV_means', 'HR_HRV_stds', stress_df, labels)
plot_data('HR_HRV_means', 'HR_HRV_RMSSDs', stress_df, labels)

plot_data('GSR_means', 'GSR_stds', stress_df, labels)
plot_data('GSR_rate_peaks', 'GSR_mean_peaks', stress_df, labels)
plot_data('GSR_mean_peaks', 'GSR_std_peaks', stress_df, labels)
```

## K-Fold Cross Validation

```python
kf = KFold(n_splits=10, random_state=1, shuffle=True)
```

# PCA

```python
# Principal Component Analysis method of dimensionality reduction
# Uses eigenvalues and eigenvectors to find new axes on which the data is most spread out

pca = PCA(n_components=2) #2-dimensional PCA
transformed = pd.DataFrame(pca.fit_transform(X))
```

```python
plt.scatter(transformed[y==2][0], transformed[y==2][1], label='2', c='red')
plt.scatter(transformed[y==3][0], transformed[y==3][1], label='3', c='orange')
plt.scatter(transformed[y==4][0], transformed[y==4][1], label='4', c='yellow')
plt.scatter(transformed[y==5][0], transformed[y==5][1], label='5', c='green')
plt.scatter(transformed[y==6][0], transformed[y==6][1], label='6', c='blue')
plt.scatter(transformed[y==7][0], transformed[y==7][1], label='7', c='purple')

plt.legend()
plt.show()
```

# LDA

```python
# Linear Discriminant Analysis project higher dimensions onto lower dimensions
# like PCA, but attempts to separate classes rather than spread the dataset

lda = LDA(n_components=2) #2-dimensional LDA
lda_transformed = pd.DataFrame(lda.fit_transform(X, y))
```

```python
plt.scatter(lda_transformed[y==2][0], lda_transformed[y==2][1], label='2', c='red')
plt.scatter(lda_transformed[y==3][0], lda_transformed[y==3][1], label='3', c='orange')
plt.scatter(lda_transformed[y==4][0], lda_transformed[y==4][1], label='4', c='yellow')
plt.scatter(lda_transformed[y==5][0], lda_transformed[y==5][1], label='5', c='green')
plt.scatter(lda_transformed[y==6][0], lda_transformed[y==6][1], label='6', c='blue')
plt.scatter(lda_transformed[y==7][0], lda_transformed[y==7][1], label='7', c='purple')
plt.legend()
plt.show()
```

# Logistic Regression

```python
# Logistic regression models the probability of a certain class being a specific label; for each label
# Statistical model that uses logistic function; extended to for multi-class classification via multinomial logistic regression

clf = sklearn.linear_model.LogisticRegression()
clf.fit(X, y)

clf_val = sklearn.linear_model.LogisticRegression()
scores = cross_val_score(clf_val, X, y, cv=kf)
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

# K-Nearest Neighbours

```python
# Non-parameteric classification method
# Output label is assigned based on a plularity vote of its neighbours
# with the object being assigned to the class most common among its k nearest neighbours

weight = 'distance' #'uniform'
n_neighbours = 5
clf = neighbors.KNeighborsClassifier(n_neighbours, weight)
clf.fit(X, y)

scores = cross_val_score(clf, X, y, cv=kf)
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
# Save the trained model
saved_model_name = "knn_model_hrv_sc_1.sav"
saved_model = pickle.dump(clf, open(saved_model_name, "wb"))
```

#  SVM

```python
# Supervised learning
# Support Vector Machine construct hyperplane(s) for classification; kernel is defined to allow separation data points via hyperplanes
# Effective in high dimensional spaces + non-linearly separable data, and where number of dimensions is greater than number of samples
# Training points are called support vectors

clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X, y)

# scores = cross_val_score(clf, X, y, scoring='accuracy', cv=kf)
# print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

# k-Means

```python
# k-Means clusters data by trying to separate samples into n groups of equal variance, minimising inertial

clf = KMeans(n_clusters=6, random_state=0).fit(X)
centroids = clf.cluster_centers_
print(clf.labels_)
```

```python
pca = PCA(n_components=2) #2-dimensional PCA
reduced_data = pca.fit_transform(X)

clf = KMeans(n_clusters=6, random_state=0).fit(reduced_data)

# Step size of the mesh. Decrease to increase the quality
h = .1     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
centroids = clf.cluster_centers_

# Put the result into a color plot
pred = pred.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(pred, interpolation="nearest", extent=(xx.min(), xx.max(), yy.min(), yy.max()), cmap=plt.cm.Paired, aspect="auto", origin="lower")
plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)

# Plot the centroids as a white X
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=169, linewidths=3, color="w", zorder=10)
plt.title("K-means clustering on the digits dataset (PCA-reduced data)\n" "Centroids are marked with white cross")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
